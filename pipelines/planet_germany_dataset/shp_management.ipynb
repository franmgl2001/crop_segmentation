```python
import geopandas as gpd

# Replace 'path/to/your/shapefile.shp' with the path to your shapefile
shapefile_path = "showcase/shp/fields17.shp"

# Load the shapefile using GeoPandas
gdf = gpd.read_file(shapefile_path)


```

```python
# Extract polygons from the 'geometry' column
polygons = gdf["geometry"]
tolat_area = 0
# Iterate through the polygons and print them
for i, polygon in enumerate(polygons):
    #print(f"Polygon {i}:", polygon)
    tolat_area += polygon.area
```
